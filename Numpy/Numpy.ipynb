```python
import numpy as np
arr=np.array([10,15,35,40])
newarr=np.diff(arr,n=2)
print(newarr)
```

NumPy LCM

```python
import numpy as np
num1=4
num2=6
x=np.lcm(num1,num2)
print(x)
```

```python
import numpy as np
arr=np.array([3,6,9])
x=np.lcm.reduce(arr)
print(x)
```

```python
import numpy as np
arr=np.arange(1,10)
x=np.lcm.reduce(arr)
print(x)
```

NumPy GCD

```python
import numpy as np
num1=6
num2=6
x=np.gcd(num1,num2)
print(x)
```

```python
import numpy as np
arr=np.array([20,8,43,23,12])
x=np.gcd.reduce(arr)
print(x)
```

Convert Degrees Into Radians

```python
import numpy as np
arr=np.array([10,20,30,40,50])
x=np.deg2rad(arr)
print(x)
```

Numpy set operations

```python
import numpy as np
arr=np.array([1,2,1,1,3,4,6,7,8,5,7])
a=np.unique(arr)
print(a)
```

Finding Union

```python
import numpy as np
arr=np.array([1,2,2,2,3,4,5,4,6,7,8,9,])
x=np.unique(arr)
print(x)
```

```python
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([2,3,6,7])
newarr=np.intersect1d(arr1,arr2,assume_unique=True)
print(newarr)
```

Finding Difference

```python
import numpy as np
set1=np.array([1,2,3,4])
set2=np.array([3,4,7,8])
newarr=np.setdiff1d(set1,set2,assume_unique=True)
print(newarr)
```

```python
import numpy as np
set1=np.array([1,2,3,4,5])
set2=np.array([6,7,4,5,8])
newarr=np.setxor1d(set1,set2,assume_unique=True)
print(newarr)
```
