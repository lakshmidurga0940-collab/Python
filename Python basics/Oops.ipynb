# Abstraction

```python
from abc import ABC,abstractmethod
class vehicle(ABC):
  @abstractmethod
  def start_engine(self):
    pass
  def stop_engine(self):
    pass
class car(vehicle):
  def start_engine(self):
    print("Car engine started")
  def stop_engine(self):
    print("Car engine stopped")
car = car()
car.start_engine()
car.stop_engine()
```

```python
from abc import ABC, abstractmethod
class Shape(ABC):
  @abstractmethod
  def calculate_area(self):
    pass

class circle(Shape):
  def __init__(self,radius):
    self.radius = radius

  def calculate_area(self):
    print(f"circle area is : {3.14 * (self.radius)**2 }")

class Rectangle(Shape):
  def __init__(self,width,height):
    self.width=width
    self.height=height

  def calculate_area(self):
    print(f"Rectangle area is:{self.width*self.height}")

cir=circle(10)
rect=Rectangle(5,10)
cir.calculate_area()
print()
rect.calculate_area()
```

3.Create an abstract class Polygon with an abstract method number_of_sides(). Implement two subclasses, Triangle and Pentagon, which return the respective number of sides. Instantiate both classes and print the number of sides.

```python
class polygon:
  @abstractmethod
  def number_of_sides(self):
    pass
class Triangle(polygon):
  def __init__(self, sides):
    self.sides=sides
def number_of_sides(self):
    print(f"Triangle side is: {self.sides}")
class Pentagon(polygon):
  def __init__(self,sides):
    self.sides=sides
  def number_of_sides(self):
    print(f"Pentagon side is:{self.sides}")
t=Triangle(3)
p=Pentagon(2)
t.number_of_sides()
print()
p.number_of_sides()
```
